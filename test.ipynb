```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

# Account
sender_email = "2026trainees86@scg.com"
password = "2025@scgc"

# Recipient
receiver_email = "2026trainees86@scg.com"

# Create email
msg = MIMEMultipart()
msg["From"] = sender_email
msg["To"] = receiver_email
msg["Subject"] = "Python Test Email"

body = """
Hello,

This is a test email sent from Python.

Regards,
Python Bot
"""

msg.attach(MIMEText(body, "plain"))

# Outlook / Microsoft 365 SMTP
smtp_server = "smtp.office365.com"
port = 587

try:
    server = smtplib.SMTP(smtp_server, port)
    server.starttls()
    server.login(sender_email, password)
    server.sendmail(sender_email, receiver_email, msg.as_string())
    server.quit()

    print("Email sent successfully!")

except Exception as e:
    print(f"Error: {e}")
```

```python
import urllib.parse
import webbrowser


def auto_redirect_to_outlook(email_to, subject, body):
    # 1. เข้ารหัสข้อความป้องกันภาษาไทยพัง
    safe_subject = urllib.parse.quote(subject)
    safe_body = urllib.parse.quote(body)

    # 2. สร้างโครงสร้าง URL ของ Outlook Web App
    outlook_base_url = "https://outlook.office.com/mail/deeplink/compose"
    final_url = f"{outlook_base_url}?to={email_to}&subject={safe_subject}&body={safe_body}"

    print("กำลังนำทางคุณไปยัง Outlook...")

    # 3. สั่งให้คอมพิวเตอร์เปิดเบราว์เซอร์เด้งไปหน้านั้นทันที (เหมือนฟิล Streamlit)
    webbrowser.open(final_url)


# --- ตัวอย่างการใช้งานเมื่อคุณเตรียมข้อมูลตารางเสร็จ ---
# พอระบบคำนวณเสร็จปุ๊บ สั่งฟังก์ชันนี้ทำงานทันที
auto_redirect_to_outlook(
    email_to="ขาดใคร_ต้องการใคร@company.com",
    subject="[สรุปผล] รายชื่อพนักงานที่ขาดข้อมูล",
    body="สวัสดีครับ,\n\nระบบได้ทำการสรุปผลเสร็จสิ้นแล้ว\nสามารถตรวจสอบและกดส่งจดหมายฉบับนี้ได้ทันทีครับ",
)
```

```python
import pandas as pd
import webbrowser

df = pd.DataFrame({
    "Customer": ["A", "B", "C"],
    "Current Debt": [1200000, 800000, 1500000],
    "Overdue": [25000, 0, 100000]
})

html = f"""
<html>
<body>

<button onclick="copyReport()">Copy Report</button>

<div id="report">
<h3>Credit Monitoring Report</h3>
{df.to_html(index=False)}
</div>

<script>
function copyReport() {{
    const report = document.getElementById('report');
    const range = document.createRange();
    range.selectNode(report);

    const selection = window.getSelection();
    selection.removeAllRanges();
    selection.addRange(range);

    document.execCommand('copy');

    alert('Copied');
}}
</script>

</body>
</html>
"""

with open("report.html", "w", encoding="utf-8") as f:
    f.write(html)

webbrowser.open("report.html")
webbrowser.open(
    "https://outlook.office.com/mail/deeplink/compose?subject=Credit%20Report"
)
```

```python
import pandas as pd
import tempfile
import webbrowser

df = pd.DataFrame({
    "Customer": ["A", "B", "C"],
    "Risk": ["High", "Medium", "Low"],
    "Overdue": [500000, 100000, 0]
})

html = f"""
<html>
<body>

<h2 style="color:red;">
Credit Monitoring Report
</h2>

{df.to_html(index=False)}

<p>
<span style="color:red;font-weight:bold;">
High Risk Customers Require Action
</span>
</p>

<script>

async function copyReport() {{

    const html = document.getElementById("report").innerHTML;

    const blob = new Blob(
        [html],
        {{type: "text/html"}}
    );

    const data = [
        new ClipboardItem({{
            "text/html": blob
        }})
    ];

    await navigator.clipboard.write(data);

    alert("Copied as Rich HTML");
}}

</script>

<div id="report">

<h2 style="color:red;">
Credit Monitoring Report
</h2>

{df.to_html(index=False)}

<p>
<span style="color:red;font-weight:bold;">
High Risk Customers Require Action
</span>
</p>

</div>

<button onclick="copyReport()">
Copy Rich HTML
</button>

</body>
</html>
"""

path = tempfile.NamedTemporaryFile(
    suffix=".html",
    delete=False
).name

with open(path, "w", encoding="utf-8") as f:
    f.write(html)

webbrowser.open(path)
webbrowser.open(
    "https://outlook.office.com/mail/deeplink/compose?subject=Credit%20Report"
)
```
