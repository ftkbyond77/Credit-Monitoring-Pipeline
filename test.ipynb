```python
import pandas as pd
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# 1. เตรียมข้อมูลชุดเดิม
mock_data = {
    'Combination_Stage': [
        '4. Critical', '4. Critical', 
        '3. Late Warning', '3. Late Warning',
        '2. Grace Period', 
        '1. Normal'
    ],
    'Customer Name': [
        'Grand Construction Co.', 'Vichai Royal Group', 
        'Global Trade Ltd.', 'Kittisak Logistics',
        'Thai Rungreung Corp', 
        'A-Plus Enterprise'
    ],
    'Overdue Amount': [640000, 300000, 20000, 10000, 20000, 10000]
}
df = pd.DataFrame(mock_data)
df = df.sort_values(by='Overdue Amount', ascending=False)

# กำหนดสีตามธีมระดับความเสี่ยง
color_map = {
    '1. Normal': '#2ecc71',
    '2. Grace Period': '#f1c40f',
    '3. Late Warning': '#e67e22',
    '4. Critical': '#e74c3c'
}

# คำนวณรายการสเตจหลักและยอดรวมเพื่อส่งให้ Root Node ของ Treemap
unique_stages = df['Combination_Stage'].unique()
stage_totals = [df[df['Combination_Stage'] == s]['Overdue Amount'].sum() for s in unique_stages]

# สร้างโครงสร้างรายการทั้งหมด (ทั้งด่านหลักและรายชื่อลูกค้า) เพื่อแมปสีให้ถูกต้องทุกบล็อก
all_labels = list(unique_stages) + [row['Combination_Stage'] + "<br>" + row['Customer Name'] for _, row in df.iterrows()]
all_parents = [""] * len(unique_stages) + list(df['Combination_Stage'])
all_values = stage_totals + list(df['Overdue Amount'])

# แมปสีให้ตรงกับประเภทด่านความเสี่ยง
mapped_colors = []
for label in all_labels:
    # ตรวจสอบว่าข้อความมีชื่อด่านอยู่หรือไม่เพื่อพ่นสีให้ถูกกลุ่ม
    matched = False
    for stage, color in color_map.items():
        if stage in label:
            mapped_colors.append(color)
            matched = True
            break
    if not matched:
        mapped_colors.append('#b2bec3')

# 2. สร้างโครง Subplots (1 แถว, 2 คอลัมน์)
fig = make_subplots(
    rows=1, cols=2, 
    specs=[[{'type': 'treemap'}, {'type': 'table'}]],
    column_widths=[0.5, 0.5],
    subplot_titles=('<b>Portfolio Proportion (Treemap)</b>', '<b>Detailed Debtor Registry (Side Table)</b>')
)

# 3. เพิ่มกราฟฝั่งซ้าย: Treemap (รวมข้อมูลเป็นก้อนเดียวกันเพื่อตัดปัญหา Bug สีสลับและพารามิเตอร์ผิด)
fig.add_trace(
    go.Treemap(
        labels=all_labels,
        parents=all_parents,
        values=all_values,
        branchvalues="total",
        marker=dict(colors=mapped_colors), # ใช้พารามิเตอร์ 'colors' ที่ถูกต้อง
        textinfo="label+value",
        hovertemplate='<b>%{label}</b><br>Amount: ฿%{value:,.0f}',
        name='Portfolio'
    ),
    row=1, col=1
)

# 4. เพิ่มข้อมูลฝั่งขวา: Side Table
row_colors = [color_map[s] for s in df['Combination_Stage']]

fig.add_trace(
    go.Table(
        header=dict(
            values=['<b>Customer Name</b>', '<b>Risk Stage</b>', '<b>Overdue Amount</b>'],
            fill_color='#2d3436',
            align='left',
            font=dict(color='white', size=12, family="Sarabun")
        ),
        cells=dict(
            values=[df['Customer Name'], df['Combination_Stage'], df['Overdue Amount'].apply(lambda x: f"฿{x:,}")],
            fill_color=['#f8f9fa', '#f8f9fa', row_colors],
            align='left',
            font=dict(size=11, family="Sarabun", color='black'),
            height=28
        )
    ),
    row=1, col=2
)

# 5. ตั้งค่า Layout ภาพรวม
fig.update_layout(
    title_text="<b>Executive Credit Combo Dashboard: Treemap & Side Table Analytics</b>",
    font=dict(family="Sarabun, sans-serif", size=13),
    margin=dict(t=80, l=20, r=20, b=20),
    height=500
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# 1. เตรียมข้อมูลชุดเดิม (ชื่ออังกฤษ สัดส่วนเอียงสุดโต่ง 94% vs 1%)
mock_data = {
    'Combination_Stage': [
        '4. Critical', '4. Critical', 
        '3. Late Warning', '3. Late Warning',
        '2. Grace Period', 
        '1. Normal'
    ],
    'Customer Name': [
        'Grand Construction Co.', 'Vichai Royal Group', 
        'Global Trade Ltd.', 'Kittisak Logistics',
        'Thai Rungreung Corp', 
        'A-Plus Enterprise'
    ],
    'Overdue Amount': [640000, 300000, 20000, 10000, 20000, 10000]
}
df = pd.DataFrame(mock_data)
df = df.sort_values(by='Overdue Amount', ascending=False)

# กำหนดสีตามสเตจความเสี่ยง
color_map = {
    '1. Normal': '#2ecc71',
    '2. Grace Period': '#f1c40f',
    '3. Late Warning': '#e67e22',
    '4. Critical': '#e74c3c'
}

# คำนวณยอดเงินรวมแยกตามสเตจเพื่อส่งให้ Funnel Chart
df_funnel = df.groupby('Combination_Stage')['Overdue Amount'].sum().reset_index()
# เรียงลำดับจากด่านแรก (Normal) ไปด่านสุดท้าย (Critical) ตามทิศทางการไหลของหนี้
stages_order = ['1. Normal', '2. Grace Period', '3. Late Warning', '4. Critical']
df_funnel['sort_idx'] = df_funnel['Combination_Stage'].apply(lambda x: stages_order.index(x))
df_funnel = df_funnel.sort_values(by='sort_idx')

# 2. สร้างโครง Subplots (1 แถว, 2 คอลัมน์) โดยฝั่งซ้ายเป็น 'funnel' ฝั่งขวาเป็น 'table'
fig = make_subplots(
    rows=1, cols=2, 
    specs=[[{'type': 'funnel'}, {'type': 'table'}]],
    column_widths=[0.5, 0.5],
    subplot_titles=('<b>Credit Pipeline (Funnel Chart)</b>', '<b>Detailed Debtor Registry (Side Table)</b>')
)

# 3. เพิ่มกราฟฝั่งซ้าย: Funnel Chart
fig.add_trace(
    go.Funnel(
        name='Portfolio Portion',
        y=df_funnel['Combination_Stage'],
        x=df_funnel['Overdue Amount'],
        textinfo="value+percent initial", # โชว์ยอดเงินจริง และสัดส่วน % เทียบกับสเตจแรกสุด
        marker=dict(
            color=[color_map[s] for s in df_funnel['Combination_Stage']],
            line=dict(width=1, color='black')
        ),
        hovertemplate='<b>% {y}</b><br>Total Amount: ฿% {x:,.0f}'
    ),
    row=1, col=1
)

# 4. เพิ่มข้อมูลฝั่งขวา: Side Table (กางรายชื่อและยอดเงินเด่นๆ ครบทุกคน)
row_colors = [color_map[s] for s in df['Combination_Stage']]

fig.add_trace(
    go.Table(
        header=dict(
            values=['<b>Customer Name</b>', '<b>Risk Stage</b>', '<b>Overdue Amount</b>'],
            fill_color='#2d3436',
            align='left',
            font=dict(color='white', size=12, family="Sarabun")
        ),
        cells=dict(
            values=[df['Customer Name'], df['Combination_Stage'], df['Overdue Amount'].apply(lambda x: f"฿{x:,}")],
            fill_color=['#f8f9fa', '#f8f9fa', row_colors],
            align='left',
            font=dict(size=11, family="Sarabun", color='black'),
            height=28
        )
    ),
    row=1, col=2
)

# 5. ตั้งค่า Layout ภาพรวมของ แดชบอร์ด
fig.update_layout(
    title_text="<b>Executive Credit Combo Dashboard: Funnel & Side Table Analytics</b>",
    font=dict(family="Sarabun, sans-serif", size=13),
    margin=dict(t=80, l=20, r=20, b=20),
    height=500
)

fig.show()
```

```python
import pandas as pd

def _dedup_for_kpi_test(df: pd.DataFrame) -> pd.DataFrame:
    if "Customer Name" in df.columns:
        dedup_col = "Customer Name"
    elif "Customer Code" in df.columns:
        dedup_col = "Customer Code"
    else:
        return df

    df = df.copy()
    df["_DEDUP_KEY"] = df[dedup_col].astype(str).str.strip().str.upper()

    group_keys = [k for k in ("YEAR", "MONTH") if k in df.columns]

    if not group_keys:
        result = df.drop_duplicates(subset=["_DEDUP_KEY"], keep="first")
        return result.drop(columns=["_DEDUP_KEY"])

    parts = []
    for _, grp in df.groupby(group_keys, sort=False):
        parts.append(grp.drop_duplicates(subset=["_DEDUP_KEY"], keep="first"))

    return pd.concat(parts, ignore_index=True).drop(columns=["_DEDUP_KEY"])


# ── รันทดสอบ ─────────────────────────────────────────────────────────────────
df = df_ava.copy()
df["YEAR"]  = 2026
df["MONTH"] = df_ava["Date_fixed"].dt.month

result = _dedup_for_kpi_test(df)

# breakdown ต่อ (YEAR, MONTH)
summary = (
    result.groupby(["YEAR", "MONTH"])
    .agg(
        customers       = ("Customer Name", "nunique"),
        CLEAN_CREDIT_SUM= ("Clean Credit", "sum"),
    )
    .reset_index()
)

print("=" * 55)
print(f"rows ก่อน dedup      : {len(df)}")
print(f"rows หลัง dedup      : {len(result)}")
print("=" * 55)
print("\n📊 breakdown per (YEAR, MONTH):")
print(summary.to_string(index=False))
print("=" * 55)
print(f"CLEAN_CREDIT.sum() รวมทุกเดือน : {result['Clean Credit'].sum():,.2f}")
print(f"unique Customer Name               : {result['Customer Name'].nunique()}")
print("=" * 55)
```
